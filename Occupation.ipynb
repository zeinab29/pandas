### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user).

### Step 3. Assign it to a variable called users.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url,sep="|",index_col="user_id")
users
```

### Step 4. Discover what is the mean age per occupation

```python
df = pd.DataFrame(users)
occupation_mean = df.groupby("occupation")["age"].mean()
occupation_mean
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
occupation_counts = df.groupby("occupation")["gender"].value_counts().unstack(fill_value=0)
occupation_counts
```

```python
occupation_counts['male_ratio'] = occupation_counts['M'] / (occupation_counts['M'] + occupation_counts['F'])
occupation_counts['male_ratio'].sort_values( ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
occupation_age = df.groupby("occupation")["age"].agg(["min","max"])
occupation_age
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
mean_age = df.groupby(["occupation","gender"])["age"].mean()
mean_age
```

### Step 8.  For each occupation present the percentage of women and men

```python
occupation_count = df.groupby("occupation")["gender"].value_counts().unstack(fill_value=0)
occupation_count["total"] = occupation_count["M"] + occupation_count["F"]
occupation_count["%M"] = (occupation_count["M"]/occupation_count['total'])*100
occupation_count["%F"] = (occupation_count["F"]/occupation_count['total'])*100
occupation_count
```
