### Introduction:

GroupBy can be summarized as Split-Apply-Combine.

### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv).

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv"
```

### Step 3. Assign it to a variable called drinks.

```python
drinks = pd.read_csv(url,sep=',')
drinks
```

```python
df = pd.DataFrame(drinks)
```

### Step 4. Which continent drinks more beer on average?

```python
avg_beer_by_continent = df.groupby("continent")["beer_servings"].mean().sort_values(ascending=False)
avg_beer_by_continent
```

```python
top_continent = avg_beer_by_continent.idxmax()
print("Continent with highest average beer consumption per capita:", top_continent)
```

### Step 5. For each continent print the statistics for wine consumption.

```python
wine_stat = df.groupby("continent")["wine_servings"].agg(['mean', 'std', 'max', 'min'])
wine_stat
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
alcohol_mean = df.groupby("continent").mean(numeric_only=True)
```

```python
alcohol_mean
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
alcohol_median = df.groupby("continent").median(numeric_only=True)
alcohol_median
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
numeric_cols = df.select_dtypes(include='number').columns
alcohol_cons = df.groupby("continent")[numeric_cols].agg(["mean","min","max"])
```
